```python
all_pos_tags = []
for i in range(len(hp_text)):
    all_pos_tags.append([])
    
ind = 0
for para in all_text:
    trees = parser.parse_sents(para)
    for tree in trees:
        treetags = tree.pos()
        rem_len = len(hp_text[ind])
        for tag in treetags[:-1]:
            all_pos_tags[ind].append(tag[1])
            if tag[0] == '"' or tag[0] == "``" or tag[0] == "''" or tag[0] == "-LRB-" or tag[0] == "-RRB-" or tag[0] == ".":
                rem_len -= 1
            else:
                rem_len -= len(tag[0])
            if rem_len <= 0:
                ind += 1
                while(ind < len(hp_text) and hp_text[ind] == "+"):
                    ind += 1
                if ind < len(hp_text):
                    rem_len = len(hp_text[ind]) + rem_len

c = [[hp_text[i],all_pos_tags[i]] for i in range(len(hp_text))]
for i in c:
    print(i)
```

```python
all_tags = {}
tn = 0
for i in all_pos_tags:
    for j in i:
        if j not in all_tags:
            all_tags[j] = tn
            tn += 1

for i in range(len(all_pos_tags)):
    for j in range(len(all_pos_tags[i])):
        all_pos_tags[i][j] = all_tags[all_pos_tags[i][j]]

print(all_tags)
c = [[hp_text[i],all_pos_tags[i]] for i in range(len(hp_text))]
for i in c:
    print(i)
```

```python
import os
one_hot_tags = []
for i in all_pos_tags:
    one_hot_tags.append([0.0]*len(all_tags))
    for j in i:
        one_hot_tags[-1][j] = 1.0

np.save(os.path.join(save_dir,"pos_tags_all"),one_hot_tags)
np.save(os.path.join(save_dir,"pos_tags_mappings_all"),all_tags)
```
