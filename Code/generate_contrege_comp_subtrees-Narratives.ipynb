```python
import benepar
from IPython.core.debugger import set_trace

parser = benepar.Parser("../brain_syntactic_representations/benepar_en2_large/")
```

```python
all_text = []
hp_text = []
para = ""

narrative_text = open("./stimuli/text/lw1new.txt","r").readlines()
for eachsent in narrative_text[0].split('.'):
    eachsent = eachsent+'.'
    for j in range(len(eachsent.strip().split())):
        if eachsent.strip().split()[j].strip()!='.':
            para += eachsent.strip().split()[j].strip() + " "
            hp_text.append(eachsent.strip().split()[j].strip())
    all_text.append(para)
    para = ""

narrative_text = open("./stimuli/text/cable_spool_fort_new.txt","r").readlines()
for eachsent in narrative_text[0].split('.'):
    eachsent = eachsent+'.'
    for j in range(len(eachsent.strip().split())):
        if eachsent.strip().split()[j].strip()!='.':
            para += eachsent.strip().split()[j].strip() + " "
            hp_text.append(eachsent.strip().split()[j].strip())
    all_text.append(para)
    para = ""
        
narrative_text = open("./stimuli/text/easy_money_new.txt","r").readlines()
for eachsent in narrative_text[0].split('.'):
    eachsent = eachsent+'.'
    for j in range(len(eachsent.strip().split())):
        if eachsent.strip().split()[j].strip()!='.':
            para += eachsent.strip().split()[j].strip() + " "
            hp_text.append(eachsent.strip().split()[j].strip())
    all_text.append(para)
    para = ""
    
narrative_text = open("./stimuli/text/black_willow_new.txt","r").readlines()
for eachsent in narrative_text[0].split('.'):
    eachsent = eachsent+'.'
    for j in range(len(eachsent.strip().split())):
        if eachsent.strip().split()[j].strip()!='.':
            para += eachsent.strip().split()[j].strip() + " "
            hp_text.append(eachsent.strip().split()[j].strip())
    all_text.append(para)
    para = ""
```

```python
print(len(hp_text), len(all_text))
```

```python
all_subtrees = []
for i in range(len(hp_text)):
    all_subtrees.append([])
    
all_lexical_prods = []

ptr = 0
for para in all_text:
    #print(ptr)
    trees = parser.parse_sents(para)
    for tree in trees:
        productions = tree.productions()
        #print(productions)
        cur_prods = []
        len_remaining = len(hp_text[ptr])
        #print(len_remaining)
        for prod in productions:
            #print(prod, prod.is_lexical(), prod.rhs()[0])
            cur_prods.append(prod)
            
            if prod.is_lexical():
                #print(len(prod.rhs()[0]), len_remaining)
                if prod.rhs()[0] == "``" or prod.rhs()[0] == "''" or prod.rhs()[0] == "-LRB-" or prod.rhs()[0] == "-RRB-":
                        len_remaining -= 1
                else:
                    len_remaining -= len(prod.rhs()[0])
                all_lexical_prods.append(prod)
            if len_remaining == 0:
                all_subtrees[ptr] = cur_prods.copy()
                ptr += 1
                if hp_text[ptr-1] == ".":
                    ptr += 1
                if ptr<len(hp_text) and hp_text[ptr-1] == ".":
                    ptr += 1
                if ptr<len(hp_text):
                    len_remaining = len(hp_text[ptr])
```

```python
import pickle
with open("all_subtrees_contrege_comp_meg.pkl","wb") as f:
    pickle.dump(all_subtrees,f)
```

```python
with open("all_subtrees_contrege_comp_meg.pkl","rb") as f:
    bck = pickle.load(f)
all_end_subtrees = []
for i in range(len(hp_text)):
    all_end_subtrees.append([])

for i in range(len(hp_text)):
    j = len(all_subtrees[i]) - 1
    prev = []
    while(j>=0):
        cur = bck[i][j]
        rhs = cur.rhs()
        if cur.is_lexical() and len(cur._rhs)>0:
            all_end_subtrees[i] = [all_subtrees[i][j]] + all_end_subtrees[i]
            cur._rhs = ()
            j -= 1
            prev.append(cur.lhs())
        else:
            checkrem = 0
            rem = 0
            for sym in rhs:
                if sym in prev:
                    rem += 1
                    prev.remove(sym)
                    checkrem = 1
                if len(prev) == 0:
                    break
            bck[i][j]._rhs = bck[i][j]._rhs[rem:]
            if checkrem == 1 and len(bck[i][j]._rhs) == 0:
                all_end_subtrees[i] = all_subtrees[i][j:]
                prev.append(all_subtrees[i][j].lhs())
            j -= 1
```

```python
sym_to_ind = {}
cnt = 1
all_subgraphs = []

word_to_ind = {}
wordnum = 10000

for subtree in all_end_subtrees:
    subgraph = []
    for edge in subtree:
        lhs = edge.lhs()
        str_lhs = str(lhs)
        rhs = edge.rhs()
        str_rhs = []
        for r in rhs:
            str_rhs.append(str(r))
        if str_lhs not in sym_to_ind:
            sym_to_ind[str_lhs] = cnt
            cnt += 1
        if edge.is_lexical():
            if id(edge) not in word_to_ind:
                word_to_ind[id(edge)] = wordnum
                wordnum += 1
            subgraph.append("{} {}".format(sym_to_ind[str_lhs],word_to_ind[id(edge)]))
        else:
            for r in str_rhs:
                if r not in sym_to_ind:
                    sym_to_ind[r] = cnt
                    cnt += 1
                subgraph.append("{} {}".format(sym_to_ind[str_lhs],sym_to_ind[r]))
    all_subgraphs.append(subgraph)
```

```python
import os
output_dir = "contrege_comp_subtrees/"

if not os.path.exists(output_dir):
    os.mkdir(output_dir)

prefix = "subGraph"
for i in range(len(all_subgraphs)):
    if len(all_subgraphs[i]) == 0:
        continue
    with open(output_dir + prefix + str(i),"w+") as f:
        for edge in all_subgraphs[i]:
            f.write(edge + "\n")
```
