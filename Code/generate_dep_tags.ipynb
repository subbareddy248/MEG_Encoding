```python
import spacy
import en_core_web_sm
from spacy import displacy
import numpy as np
from IPython.core.debugger import set_trace
import os

save_dir = "features"
if not os.path.exists(save_dir):
    os.mkdir(save_dir)
    
nlp = en_core_web_sm.load()
```

```python
all_text = []
hp_text = []
para = ""
```

```python
narrative_text = open("./stimuli/text/lw1new.txt","r").readlines()
for eachsent in narrative_text[0].split('.'):
    all_text.append(eachsent.strip()+'.')
    for j in range(len(eachsent.strip().split())):
        #if eachsent.strip().split()[j].strip() not in [',','?','!','-','"',';']:
        hp_text.append(eachsent.strip().split()[j].strip())
print(all_text)
```

```python
narrative_text = open("./stimuli/text/black_willow_new.txt","r").readlines()
for eachsent in narrative_text[0].split('.'):
    all_text.append(eachsent.strip()+'.')
    for j in range(len(eachsent.strip().split())):
        #if eachsent.strip().split()[j].strip() not in [',','?','!','-','"',';']:
        hp_text.append(eachsent.strip().split()[j].strip())
```

```python
t = nlp(all_text[0])
displacy.render(t, style='dep')
```

```python
all_dep_tags = []
for i in range(len(hp_text)):
    all_dep_tags.append([])

ind = 0
rem_len = len(hp_text[ind])
for para in all_text:
    proc = nlp(para)
    for word in proc:
        if word.text!='.':
            all_dep_tags[ind].append(word.dep_)
            rem_len -= len(word.text)
            if rem_len <= 0:
                ind += 1
                while(ind < len(hp_text) and hp_text[ind] == "+"):
                    ind += 1
                if ind < len(hp_text):
                    rem_len = len(hp_text[ind]) + rem_len
c = [[hp_text[i],all_dep_tags[i]] for i in range(len(hp_text))]
for i in c:
    print(i)
```

```python
all_tags = {}
tn = 0
for i in all_dep_tags:
    for j in i:
        if j not in all_tags:
            all_tags[j] = tn
            tn += 1

for i in range(len(all_dep_tags)):
    for j in range(len(all_dep_tags[i])):
        all_dep_tags[i][j] = all_tags[all_dep_tags[i][j]]

print(all_tags)
c = [[hp_text[i],all_dep_tags[i]] for i in range(len(hp_text))]
for i in c:
    print(i)
```

```python
import os
one_hot_tags = []
for i in all_dep_tags:
    one_hot_tags.append([0.0]*len(all_tags))
    for j in i:
        one_hot_tags[-1][j] = 1.0

np.save(os.path.join(save_dir, "dep_tags_all"),one_hot_tags)
np.save(os.path.join(save_dir, "dep_tags_mappings_all"),all_tags)
```
